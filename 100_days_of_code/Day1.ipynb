```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
# plots appear inline and are stored within the notebook
%matplotlib inline
```

```python
x = np.linspace(-10, 10, 500)
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.grid(color='grey')
plt.title('Hello Plot')
plt.show()
```

```python
# make a 3d line plot
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax3d = fig.gca(projection='3d')
t = np.linspace(0, 20, 500)
x = np.sin(t)
y = np.cos(t)
ax3d.plot(x, y, t)
ax3d.w_xaxis.set_ticks([-1, 0, 1])
ax3d.set_xlabel('x')
ax3d.w_yaxis.set_ticks([-1, 0, 1])
ax3d.set_ylabel('y')
ax3d.set_zlabel('time (s)');
```
